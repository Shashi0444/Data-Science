```python
# Let's drop Customer ID since it has no meaning here 
dataSet.drop("CUST_ID", axis = 1, inplace= True)
```

#### Missing values

```python
# check the null values
dataSet.isnull().sum()
```

CREDIT_LIMIT - 1 missing value

MINIMUM_PAYMENTS - 313 missing value                  

```python
#Treating the missing values
dataSet.loc[(dataSet['CREDIT_LIMIT'].isnull())] = dataSet['CREDIT_LIMIT'].mean()
```

```python
#Treating the missing values
dataSet.loc[(dataSet['MINIMUM_PAYMENTS'].isnull())] = dataSet['MINIMUM_PAYMENTS'].mean()
```

```python
#missing values got treated
dataSet.isnull().sum()
```

#### Outliers treatment

```python
def boxplt(d):
    fig, ax = plt.subplots(figsize = (15,10))
    sns.boxplot(data = d,width=.5,ax=ax,fliersize=3)
```

```python
boxplt(dataSet)
```

```python
# Use z scores to remove outliers
from scipy import stats
def calZScore(d):
    z = np.abs(stats.zscore(d))
    dataSet1 = dataSet[(z < 3).all(axis=1)]
```

```python
calZScore(dataSet)
```

```python
boxplt(dataSet1)
```

```python
dataSet1.shape
```

### Univariate Analysis

```python
def plotdist(d, e):
    for i, col in enumerate(d.columns):
        if(col != e):
            plt.figure()
            sns.distplot(d[col])
```

### Scalling

```python
from sklearn.preprocessing import StandardScaler, normalize
```

```python
# Let's scale the data before we begin modelling
scaler = StandardScaler()
dataSet_scaled = scaler.fit_transform(dataSet1)
```

```python
dataSet_scaled.view()
```

### Model Building

```python
from sklearn.cluster import KMeans
```

```python
# Obtain optimal number of clusters
def getOptimalNumOfCluster(d):
    scores_1 = []
    range_values = range(1, 20)
    for i in range_values:
      kmeans = KMeans(n_clusters = i)
      kmeans.fit(d)
      scores_1.append(kmeans.inertia_) 
    plt.plot(scores_1, 'bx-')
    plt.title('Finding the right number of clusters')
    plt.xlabel('Clusters')
    plt.ylabel('Scores') 
    plt.show()
```

```python
getOptimalNumOfCluster(dataSet_scaled)
```

Inertia is WCSS -> it will decrease with increase in number of cluster. We will try to get the optimal number number of cluster from the elbow of the graph(from where the rate of fall in wcss has decressed significanlty.). Idea here is wcss is not decresing much but my clusters are increasing. So we decided to stop on some point in btw.

Model with least wcss is the best but then we can not take so many cluster. So we tried to get an optimal point.

Will take the optimal number of cluster = 4

```python
# Build the KMeans model
kmeans = KMeans(4)
kmeans.fit(dataSet_scaled)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
#including label in dataset
dataSet1["labels"] = labels
```

```python
dataSet1
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [dataSet.columns])
cluster_centers 
```

These are scaled column and so are not making much sense. So we need to inverse transfom them into there original scale

```python
pd.crosstab(index=dataSet1['labels'],columns='count')
```

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers, columns = [dataSet.columns])
cluster_centers
```

CASH_ADVANCE : cluster 0, all people have an avergae of 2962.705054. This is very different from other clusters.
    This means this is the category of people who are likely to take advance cash from ATM using credit card. So for campaing targeting such customer, cluster 0 should be considered.
    
cluster 1 has highest value for PURCHASES,ONEOFF_PURCHASES,INSTALLMENTS_PURCHASES, ONEOFF_PURCHASES_FREQUENCY, PURCHASES_TRX, CREDIT_LIMIT, PAYMENTS. They are heavy user of credit card. There full payment frequncy is also good.

PRC_FULL_PAYMENT has high value for cluster 1 and cluster 3 while cluster 0 and cluster 2 are on lower center.

cluster 2 and 3 are the largest clusters.
